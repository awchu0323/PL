```python

import re 
new_content = []
for i in data["內文"]:
    a = re.sub("\n","",i)
    a = re.sub("\u3000","",a)
    a = re.sub(" ","",a)
    a = re.sub("\uf67e","",a)
    a = re.sub("\uf67d","",a)
    a = re.sub("\uf67c","",a)
    new_content.append(a)
data["內文"] = new_content
data["內文"][0]
```

```python
date = []
for i in range(len(data)):
    a = data["內文"][i]
    a1 = re.findall(r"中華民國\d+年\d+月\d+日",a)
    a2 = a1[0].split("國")[1]
    date.append(a2)

data["判決時間"] = date
data.head(5)
```

```python
data.to_csv("最高法院_2022.csv",encoding="utf-8")
```
