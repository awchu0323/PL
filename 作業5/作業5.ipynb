```python
import pandas as pd
from collections import Counter
data = pd.read_csv("11101-11103毒品案件資料 .csv")
data
```

```python
data['seg'] = ""
data['seg1'] = ""
data.head()
```

```python
index = 0
for documents in data['oc_addr']:
    seg_list = jieba.analyse.extract_tags(documents, 1500)
    data['seg'].iloc[index] = seg_list
    index += 1
```
